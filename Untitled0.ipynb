```python
import leveldb
import json
import numpy as np
import pandas as pd
```

```python
db = leveldb.LevelDB('build/trades.leveldb')
```

```python
ticks = pd.DataFrame.from_dict(map(json.loads, [it[1] for it in db.RangeIter()]))
```

```python
trades = filter(lambda x: x['price_currency'] == 'USD', ticks[ticks['channel_name'] == "trade.BTC"]['trade'])
df = pd.DataFrame(map(lambda x: x['price'], trades), columns=['p'])
```

```python
rets = ((df / df.shift(1)) - 1.0) * 100
rets['dp'] = rets['p'].shift(1)
rets = rets.fillna(0.0)
```

```python
s = rets['p'] * rets['dp']
```

```python
w = np.sign(s) * -1
```

```python
plot(cumsum(w * (rets['p'].shift(1).fillna(0.0))))
```
